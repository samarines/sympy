# As√≠ntotas en funciones racionales

```python
from sympy import *
x, y = symbols('x,y')
from sympy.calculus.util import continuous_domain
```

## 224

```python
f = (x-2)/(x+3)
```

```python
f
```

```python
plot(f)
```

```python
f.subs(x,0) ## intersecci√≥n del eje y
```

```python
solve(f,x) ## intersecci√≥n con el eje x
```

```python
plot(f, ylim=(-5,5))
```

```python
limit(f,x,oo)
```

Los valores candidatos a existencia de as√≠ntota vertical son los siguientes:

- Valores que anulan alg√∫n denominador de la funci√≥n
        
- Extremos de intervalos del dominio que no pertenezcan al propio dominio


```python
continuous_domain(f, x, S.Reals) # C√°lculo del dominio
```

```python
limit(f,x,-3)
```
